1. Add the current date to the text file today.txt as a string.

```python
import datetime
import traceback
try:
    f=open("today.txt","w")
    f.write(str(datetime.date.today())) 
    f.close()
except:
    traceback.print_exc()      
else:
    print("current date added to today.txt")
```

2. Read the text file today.txt into the string today_string

```python
import traceback

try :
    file=open("today.txt","r")
    today_string=str(file.read())

except: 
    traceback.print_exc()    
else:
        
    print("Successfull fetched today date",today_string)
```

6. Use multiprocessing to create three separate processes. Make each one wait a random number of seconds between one and five, print the current time, and then exit.

```python
# importing the multiprocessing module
import multiprocessing
import datetime
import time
import os
import random

def printCurrentTime(process_nm):
    wait_time=random.randint(1,5)
    print("Time from process before wait",process_nm,str(os. getpid()),str(datetime.datetime.now()))
    time.sleep(wait_time)
    print("Time from process after wait",process_nm,str(os. getpid()),str(datetime.datetime.now()))

if __name__=="__main__":
    print("Main Process PID", os. getpid() )
    p1 = multiprocessing.Process(target=printCurrentTime,args=("p1",))
    p2 = multiprocessing.Process(target=printCurrentTime,args=("p2",))
    p3 = multiprocessing.Process(target=printCurrentTime,args=("p3",))
    p4 = multiprocessing.Process(target=printCurrentTime,args=("p4",))

    p1.start()   
    p2.start()
    p3.start()
    p4.start()
```

9. When will you be (or when were you) 10,000 days old?

```python
import datetime
from dateutil.relativedelta import relativedelta
d=datetime.datetime.strptime('1990-10-22','%Y-%m-%d').date()
new_date = d + relativedelta(days=10000)
print(new_date)
```

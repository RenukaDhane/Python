```python
1.Create a function that takes a list and returns a new list containing only prime numbers.

Examples

filter_primes([7, 9, 3, 9, 10, 11, 27]) ➞ [7, 3, 11]

filter_primes([10007, 1009, 1007, 27, 147, 77, 1001, 70]) ➞ [10007, 1009]

filter_primes([1009, 10, 10, 10, 3, 33, 9, 4, 1, 61, 63, 69, 1087, 1091, 1093, 1097]) ➞ [1009, 3, 61, 1087, 1091, 1093, 1097]
```

```python
import math
def filter_primes(num_list):
    primes=[]
    for num in num_list:
        if num==1:
            continue
        prime_flag=True    
        for x in range(2,int(math.sqrt(num))+1):
            if num%x==0:
                prime_flag=False
                break
                    
        if  prime_flag==True:
            primes.append(num)
            
    return primes
                       

print(filter_primes([7, 9, 3, 9, 10, 11, 27]))   
print(filter_primes([10007, 1009, 1007, 27, 147, 77, 1001, 70]))
print(filter_primes([1009, 10, 10, 10, 3, 33, 9, 4, 1, 61, 63, 69, 1087, 1091, 1093, 1097])) 
                   
                
```

```python
2.Once a water balloon pops, is soaks the area around it. The ground gets drier the further away you travel from the balloon.

The effect of a water balloon popping can be modeled using a list. 
Create a function that takes a list which takes the pre-pop state 
and returns the state after the balloon is popped. 
The pre-pop state will contain at most a single balloon, whose size is represented by the only non-zero element.

Examples

pop([0, 0, 0, 0, 4, 0, 0, 0, 0]) ➞ [0, 1, 2, 3, 4, 3, 2, 1, 0]

pop([0, 0, 0, 3, 0, 0, 0]) ➞ [0, 1, 2, 3, 2, 1, 0]

pop([0, 0, 2, 0, 0]) ➞ [0, 1, 2, 1, 0]

pop([0]) ➞ [0]
```

```python
def pop(num_list):
    output_list=[]
    counter=0
    ripple_flag=False
    for num in num_list:
        if num==0 and ripple_flag==False :
            output_list.append(counter)
            counter+=1
            
        elif num!=0:
            output_list.append(num)
            ripple_flag=True
            
        elif num==0 and ripple_flag==True :
            counter-=1
            output_list.append(counter)
            
            
    return output_list


print(pop([0, 0, 0, 0, 4, 0, 0, 0, 0]) )
print(pop([0, 0, 0, 3, 0, 0, 0]))
print(pop([0, 0, 2, 0, 0]))
print(pop([0]))
```

```python
3."Loves me, loves me not" is a traditional game in which a person plucks off all the petals
of a flower one by one, saying the phrase "Loves me" and "Loves me not" 
when determining whether the one that they love, loves them back.

Given a number of petals, return a string which repeats the phrases "Loves me" and "Loves me not"
for every alternating petal, and return the last phrase in all caps. Remember to put a comma and space between phrases.

Examples

loves_me(3) ➞ "Loves me, Loves me not, LOVES ME"

loves_me(6) ➞ "Loves me, Loves me not, Loves me, Loves me not, Loves me, LOVES ME NOT"

loves_me(1) ➞ "LOVES ME"
```

```python
def loves_me(num):
    output_str=""
    for i in range(1,num+1):
        if i%2!=0:
            if i==num:
                output_str+="Loves me".upper()
            else:
                output_str+="Loves me"
                output_str+=","
            
        else:
                if i==num:
                    output_str+="Loves me not".upper()
                else:
                    output_str+="Loves me not"
                    output_str+=","
                    
    return output_str


print(loves_me(3) )
print(loves_me(6) )
print(loves_me(1))
```

```python
4. Write a function that sorts each string in a list by the letter in alphabetic ascending order (a-z).

Examples

sort_by_letter(["932c", "832u32", "2344b"])
➞ ["2344b", "932c", "832u32"]

sort_by_letter(["99a", "78b", "c2345", "11d"])
➞ ["99a", "78b", "c2345", "11d"]

sort_by_letter(["572z", "5y5", "304q2"])
➞ ["304q2", "5y5", "572z"]

sort_by_letter([])
➞ []
```

```python
import re
def sort_by_letter(word_list):
    sorted_list=[]
    dict_words=dict()
    for word in word_list:
        letter= (re.search(r'[a-z]',word)).group()
        dict_words[letter]=word
    
    for key in sorted(dict_words.keys()):
        sorted_list.append(dict_words[key])
    
    return sorted_list



print(sort_by_letter(["932c", "832u32", "2344b"]))
print(sort_by_letter(["99a", "78b", "c2345", "11d"]))
print(sort_by_letter(["572z", "5y5", "304q2"]))
print(sort_by_letter([]))
```

5. There are three cups on a table, at positions A, B, and C. At the start, there is a ball hidden under the cup at position B.


However, I perform several swaps on the cups, which is notated as two letters. For example, if I swap the cups at positions A and B, I could notate this as AB or BA.

Create a function that returns the letter position that the ball is at, once I finish swapping the cups. 
The swaps will be given to you as a list.

Example

cup_swapping(["AB", "CA", "AB"]) ➞ "C"

# Ball begins at position B.
# Cups A and B swap, so the ball is at position A.
# Cups C and A swap, so the ball is at position C.
# Cups A and B swap, but the ball is at position C, so it doesn't move.


```python
def cup_swapping(swap_list):
    cup_with_ball="B"
    for swap in swap_list:
        if cup_with_ball in swap:
            cup_with_ball=swap.replace(cup_with_ball,'')
    
    return cup_with_ball


cup_swapping(["AB", "CA", "AB"])
```
